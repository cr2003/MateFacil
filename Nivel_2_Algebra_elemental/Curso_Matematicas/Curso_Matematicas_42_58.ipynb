```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, b, c, d, m, n, o, p, q, r, s, t, v, x, y, z = sp.symbols(
    "a, b, c, d, m, n, o, p, q, r, s, t, v, x, y, z", real=True
)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
spacing_implies = r"\, \implies \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

```python
def latex_factors(numero):
    """Devuelve en formato latex los factores del numero"""
    factores = sp.factorint(numero)

    # LaTex format
    partes = []
    for primo, exp in factores.items():
        if exp == 1:
            partes.append(str(primo))
        else:
            partes.append(f"{primo}^{{{exp}}}")
    # Unimos las partes con el símmbolo de multiplicacion LaTex (\cdot)
    latex_factors = r" \cdot ".join(partes)
    return latex_factors
```

### ¿Qué es una Ecuación? Tipos de ecuaciones y propiedades
#### Ejemplo

$$ 5x + 2 = 7 - 2x $$

### Tipos de ecuaciones

* Ecuaciones algebraicas

    * Primer Grado (Lineales)

       $ 5x + 2 = 7 - 2x $

    * Segundo Grado (Cuadráticas)

       $ 2x^2 -6 = 3x -2 $

    * TercerGrado (Cúbicas)

       $ x^3 -6x = x^2 + 8 $

    * Grado "n"

       $ ax^n + ... = ... $

    * Racionales

       $ \frac{3x^3 -2}{5x +4} = \frac{x^2}{4 -x} $

    * Irracionales

       $ \sqrt{2x - 1} = 4x $

* Ecuaciones Trascendentes

    * Ecuaciones exponenciales

       $ 2^{x+1} = 32 $

    * Ecuaciones logarítmicas

       $ \log_5(2x + 3) = 3 $

    * Ecuaciones Trigonométricas

       $ \cos(3x +2) = \frac{1}{2} $

### Sistemas de ecuaciones

$$
\begin{cases} 
2x + y = 1 \\ 
3x - 2y = 4 
\end{cases}
$$


### Propiedades de las ecuaciones

1. Reflexiva

$$x = x$$

$$5= 5$$

2. Simétrica

$$\text{Si } a = b \text{ entonces } b = a$$

$$ 3 = 2x -1 \implies 2x -1 = 3$$

3. Transitiva

$$\text{Si } a = b \text{ y } b = c \text{ entonces } a = c$$

$$ x = 3y  \text{ y } 3y = 7 \text{ entonces } x = 7$$

#### Problemas de mezclas

¿Cuántos litros de alcohol al 10% deben añadirse a 20 litros de otra solución al 40% de alcohol para obtener finalmente una solución al 30%?

<img src="../../pics/mezclas.png" width=600> 

$$ \frac{40}{100}(20) +  \frac{10}{100}(x) = \frac{30}{100}(20 + x)$$

```python
expr_raw = r" \frac{40}{100}(20) +  \frac{10}{100}(x) = \frac{30}{100}(20 + x)   "
eq = sp.Eq(
    sp.Rational(40, 100) * 20 + sp.Rational(10, 100) * x,
    sp.Rational(30, 100) * (20 + x),
)
solution = sp.solve(eq, x)[0]
display(Math(rf"{expr_raw} {spacing_implies} x = {sp.latex(solution)}"))
print(f"Alcohol que deben añadirse: {solution} litros")
```
