```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, b, c, d, m, n, p, q, r, s, t, v, x, y, z = sp.symbols(
    "a, b, c, d, m, n, p, q, r, s, t, v, x, y, z", real=True
)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

```python
def latex_factors(numero):
    """Devuelve en formato latex los factores del numero"""
    factores = sp.factorint(numero)

    # LaTex format
    partes = []
    for primo, exp in factores.items():
        if exp == 1:
            partes.append(str(primo))
        else:
            partes.append(f"{primo}^{{{exp}}}")
    # Unimos las partes con el símmbolo de multiplicacion LaTex (\cdot)
    latex_factors = r" \cdot ".join(partes)
    return latex_factors
```
