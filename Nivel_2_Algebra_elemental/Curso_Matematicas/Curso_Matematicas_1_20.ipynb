```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, b, c, d, m, n, p, q, r, s, t, v, x, y, z = sp.symbols(
    "a, b, c, d, m, n, p, q, r, s, t, v, x, y, z", real=True
)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

## Video 16

###  Cómo Sumar y Restar Polinomios

$$  5x -2y + 4z$$
$$  4x -4y - z$$


```python
expr_raw = r" 5x -2y + 4z +  4x -4y - z"
pol1 = 5 * x - 2 * y + 4 * z
pol2 = 4 * x - 4 * y - z
result = pol1 + pol2

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

###  Restar Polinomios

$$  2xy -3xz + 2yz $$
$$  xz -xy + yz $$


```python
expr_raw = r"2xy -3xz + 2yz - (xz -xy + yz)"
pol1 = 2 * x * y - 3 * x * z + 2 * y * z
pol2 = x * z - x * y + y * z
result = pol1 - pol2

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

###  Cómo Dividir Polinomios Binomios y Monomios
$$ \frac{x^m}{x^n} = x^{m - n} $$

$$ \frac{x^5}{x^2} $$

```python
expr_raw = r" \frac{x^5}{x^2}   "
result = x**5 / x**2
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

$$ \frac{x^{10} y^7}{x^5 y}  $$

```python
expr_raw = r" \frac{x^{10} y^7}{x^5 y}   "
result = (x**10 * y**7) / (x**5 * y)
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

$$ \frac{-10a^{5} b^6 c^2}{2a^4 b^6}  $$

```python
expr_raw = r" \frac{-10a^{5} b^6 c^2}{2a^4 b^6}   "
result = (-10 * a**5 * b**6 * c**2) / (2 * a**4 * b**6)
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

$$ \frac{m^2}{m^5} = m^{2 - 5} = m^{-3} = \frac{1}{m^3} $$

$$ \frac{x^3 y^{10}}{x^5 y^6} $$

```python
expr_raw = r" \frac{x^3 y^{10}}{x^5 y^6}   "
result = (x**3 * y**10) / (x**5 * y**6)
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

$$ \frac{12m^2 n^3 p}{10m^2 n^2 p2} $$

```python
expr_raw = r" \frac{12m^2 n^3 p}{10m^2 n^2 p2}   "
result = (sp.Integer(12) * m**2 * n**3 * p) / (sp.Integer(10) * m**2 * n**2 * p**2)
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

$$ \frac{9x^3 - 12x^2 + 6x^4}{3x^2} $$

```python
expr_raw = r" \frac{9x^3 - 12x^2 + 6x^4}{3x^2}   "
result = (sp.Integer(9) * x**3 - sp.Integer(12) * x**2 + sp.Integer(6) * x**4) / (
    sp.Integer(3) * x**2
)
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result.expand())}"))
```

$$ \frac{x^{-2}}{x^7} $$
