#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

```python
expr_raw = r"\sqrt{819.442}"
expr_sp = sp.sqrt(819.442)
# spacing = r"\quad = \qquad"  # Ajusta la cantidad de \qquad según necesites
spacing = r"\,= \,"
expr_sp_format = f"{expr_sp:.3f}"
# expr_sp_format = f"{expr_sp}"

display(Math(rf"{expr_raw} {spacing} {sp.latex(expr_sp_format)} "))
```

```python
def latex_factors(numero):
    """Devuelve en formato latex los factores del numero"""
    factores = sp.factorint(numero)

    # LaTex format
    partes = []
    for primo, exp in factores.items():
        if exp == 1:
            partes.append(str(primo))
        else:
            partes.append(f"{primo}^{{{exp}}}")
    # Unimos las partes con el símmbolo de multiplicacion LaTex (\cdot)
    latex_factors = r" \cdot ".join(partes)
    return latex_factors


# Use
expr_raw = r" factores \, primos: 2520 "
expr = 2520
result = latex_factors(expr)
display(Math(rf"{expr_raw} {spacing_equal} {result}"))
```

### Convertir una fracción impropia en una fracción mixta

```python
def frac_to_frac_mixta(frac):
    # Extraemos p y q
    p = frac.p
    q = frac.q

    # Calculamos la parte mixta
    entero = p // q
    residuo = p % q

    # Formateamos para mostrar
    if residuo == 0:
        resultado = rf"{entero}"
    else:
        resultado = rf"{entero} \frac{{{residuo}}}{{{q}}}"
    return resultado


expr_raw = r"\frac{5}{2}"
frac = sp.Rational(5, 2)
resultado = frac_to_frac_mixta(frac)
display(Math(rf"{expr_raw} {spacing} {resultado}"))
```

```python
expr_raw = r"\frac{6}{2}"
frac = sp.Rational(6, 2)
resultado = frac_to_frac_mixta(frac)
display(Math(rf"{expr_raw} {spacing} {resultado}"))
```

### Convertir una fracción mixta en una fracción impropia 

```python
def frac_mixta_to_frac(entero, frac):
    """
    Recibe un entero y una fraccion y devuelve una fraccion impropia.
    frac.p es el numerador de la fraccion
    frac.q es el denominador de la fraccion
    """
    numerador = (frac.q * entero) + frac.p
    denominador = frac.q
    return sp.Rational(numerador, denominador)
```

```python
expr_raw = r"5 \frac{1}{2}"
entero = 5
frac_mix = sp.Rational(1, 2)
frac = frac_mixta_to_frac(entero, frac_mix)
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(frac)}"))
```

### Convertir números decimales a fracciones

```python
def decimal_a_fraccion(decimal_float):
    # 1. Convertimos el float a string para contar cuántos decimales tiene
    decimal_str = str(decimal_float).split(".")[-1]
    num_decimales = len(decimal_str)

    # 2. Calculamos el numerador y denominador de la fracción no simplificada
    numerador_original = int(round(decimal_float * (10**num_decimales)))
    denominador_original = 10**num_decimales

    # 3. Creamos las dos fracciones en SymPy
    # Usamos sp.sympify con la cadena exacta para evitar que se autoevalúe/simplifique
    frac_original = sp.sympify(
        f"{numerador_original}/{denominador_original}", evaluate=False
    )
    frac_simplificada = sp.Rational(numerador_original, denominador_original)

    # 4. Construimos la cadena en formato LaTeX
    latex_expr = (
        rf"{decimal_float} = "
        rf"{sp.latex(frac_original)} = "
        rf"{sp.latex(frac_simplificada)}"
    )

    # 5. Mostramos el resultado estético en el Notebook
    display(Math(latex_expr))


decimal_a_fraccion(3.44)
```

### Formateo de salida para números largos con blanco como separador

```python
def fmt(n):
    return f"{int(n):,}".replace(
        ",", r"\;"
    )  # LaTeX usa {,} como separador, cambiar por un espacio en blanco
```

### Función para convertir un numero decimal a notación cientifica, expresando la precisión requerida

```python
def to_latex_sci(num, decimals=2):
    mantissa, exp = f"{num:.{decimals}e}".split("e")
    exp = int(exp)
    return rf"{mantissa} \times 10^{{{exp}}}"
```

```python
num = 23_300_766_192_040
num_sci_latex = to_latex_sci(num)

spacing = r"\, \approx \,"
display(Math(rf"{fmt(num)} {spacing}  {num_sci_latex}"))
```

### Función para adaptar el exponente de notación científica a uno predeterminado

```python
def adaptar_exponente_latex(numero, exponente_objetivo):
    # 1. Calculamos el nuevo coeficiente usando la precisión de SymPy
    nuevo_coeficiente = numero / (10**exponente_objetivo)
    coef_float = float(nuevo_coeficiente)

    # 2. Si es entero, limpiamos el .0
    if coef_float.is_integer():
        coef_str = str(int(coef_float))
    else:
        # LA CLAVE: Forzamos a Python a redondear directamente en la cadena de texto.
        # Usamos :.12f para limpiar la basura binaria, y luego quitamos los ceros sobrantes.
        coef_str = f"{coef_float:.12f}".rstrip("0").rstrip(".")

    return rf"{coef_str} \times 10^{{{exponente_objetivo}}}"
```

```python
expr_raw = r"12.51 \times 10^{-5}"
num = sp.Float(12.51e-5)
res_1_neg3 = adaptar_exponente_latex(num, -3)
res_2_neg8 = adaptar_exponente_latex(num, -8)
res_3_exp2 = adaptar_exponente_latex(num, 2)

display(
    Math(
        rf"{expr_raw} {spacing}  {res_1_neg3} {spacing}  {(res_2_neg8)} {spacing}  {res_3_exp2}"
    )
)
```

### Regla de tres compuesta

Se emplean 12 hombres durante 6 días para cavar una zanja de 30m de largo, 8m de ancho y 4m de alto, trabajando 6 horas diarias. Si se emplea el doble número de hombres durante 5 días, para cavar otra zanja de 20m de largo, 12m de ancho y 3m de alto, ¿cuántas horas diarias deben trabajar?

* Dos cantidades son `directamente proporcionales`, si al aumentar una aumenta también la otra $\mp$

    * Si aumenta el largo de la zanja, aumentará el número de horas de trabajo

* Dos cantidades son `inversamente proporcionales`, si al aumentar una disminuye la otra $\pm$

    * Si aumenta el número de hombres, disminuirá el número de horas de trabajo

#### Solución:

* Paso 1. Escribimos las cantidades, de forma ordenada

* Paso 2. Colocar los signos

<br>

<div align="center">

| + | + | - | - | - | + |
| :---: | :---: | :---: | :---: | :---: | :---: |
| **Hombres** | **Días** | **Largo** | **Ancho** | **Alto** | **Horas** |
| 12 | 6 | 30 | 8 | 4 | 6 |
| 24 | 5 | 20 | 12 | 3 | $x$ |
| - | - | + | + | + | - |

</div>

* Paso 3. Multiplicar todos los + y dividir entre todos los -

$$x = \frac{12 \times 6 \times 20 \times 12 \times 3 \times 6}{24 \times 5 \times 30 \times 8 \times 4}$$



```python
spacing = r"\, = \,"
expr_raw = r"\frac{12 \times 6 \times 20 \times 12 \times 3 \times 6}{24 \times 5 \times 30 \times 8 \times 4}"
x = expr_sp = (12 * 6 * 20 * 12 * 3 * 6) / (24 * 5 * 30 * 8 * 4)

display(Math(rf"x = {expr_raw} {spacing} {x} \text{{ horas al día}}"))
```

```python
# Convertimos a string y separamos por el punto decimal
str_entera, str_decimal = str(x).split(".")

horas = int(str_entera)
parte_decimal = float(f"0.{str_decimal}")
minutos = int(parte_decimal * 60)

display(
    Math(
        rf" {x} \text{{ horas al día}} {spacing} {horas} \text{{ horas con }} {minutos} \text{{ minutos al día}}"
    )
)
```
