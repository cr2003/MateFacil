```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, b, c, d, m, n, p, q, r, s, t, v, x, y, z = sp.symbols(
    "a, b, c, d, m, n, p, q, r, s, t, v, x, y, z", real=True
)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

## Video 64
### Leyes de exponentes
$$ 
a^n = a \times a \times \cdots \times a \;\text(\,n \,veces)
$$

$$3^2 = 3 \times 3 = 9$$

$$4^3 = 4 \times 4 \times 4= 64$$

$$1^{10} = 1$$

$$0^{100} = 0$$

$$3^1 = 3$$

$$a^1 = a$$

$$ 2^3 \times 2^4 = 2^5$$

$$ a^m \times a^n = a^{m+n}$$

$$ \frac{a^m}{a^n} = a^{m-n}$$

$$a^0 = 1$$

$$ a^{-n} = \frac{1}{a^n}$$

$$ (a^m)^n = a^{mn}$$

$$ (\sqrt[n]{a})^n = a$$

$$ a^{1/n} = \sqrt[n]{a}$$

$$ a^{m/n} = \sqrt[n]{a^m}$$

## Video 79
### Clasificación de números: Naturales, Enteros, Racionales, Irracionales, Reales, Complejos

#### Naturales
$$ 
1, 2, 3, 4, 5, 6, \dots \;\; \in \mathbb{N}
$$


#### Enteros
$$ 
\dots, -3, -2, -1, 0, 1, 2, 3, 4, 5, 6, \dots \;\; \in \mathbb{Z}
$$


#### Racionales
$$ 
\frac{a}{b}  \;\; \text(a, \; b \; son\;  entereos\; ) \in \mathbb{Q}
$$

Tienen notación decimal periódica


#### Irracionales

Tienen notación decimal no periódica

No se pueden representar como $\frac{a}{b}$ (a, b, enteros)

$$ 
\pi, \sqrt{2}, \sqrt{10}, \sqrt[3]{7}, e  \in \mathbb{I}
$$




#### Reales

Conjunto formado por los racionales y los irracionales

$$
\in \mathbb{R}
$$







#### Complejos

$$
i^2 = -1 
$$

$$
 a + bi \;\; \in \mathbb{C} \;\text(a, \; b \; son\;  reales \; ) 
$$

$$
 x^2 = -9 \implies x=3i
$$
