```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, b, c, d, m, n, p, q, r, s, t, v, x, y, z = sp.symbols(
    "a, b, c, d, m, n, p, q, r, s, t, v, x, y, z", real=True
)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

## Video 52
### Leyes de los radicales

$$ 
\sqrt[n]{1} = 1
$$

$$ 
\sqrt[n]{0} = 0
$$

```python
expr_raw = r"\sqrt[5]{96} "
result = sp.root(sp.Integer(96), 5)

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

#### Ley de cancelación

$$ 
(\sqrt[n]{a})^n = a
$$

```python
expr_raw = r"(\sqrt{4})^2 "
result = (sp.sqrt(sp.Integer(4))) ** 2

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

#### Raíz de una multiplicación (o producto)
$$ 
\sqrt[n]{ab} = \sqrt[n]{a} \sqrt[n]{b}
$$

$$
\sqrt{900} = \sqrt{9 \cdot 100} = \sqrt{9} \cdot \sqrt{100} = 3 \cdot 10 = 30
$$

```python
expr_raw = r"(\sqrt[3]{8 \cdot 27}) "
result = sp.root(sp.Integer(8) * sp.Integer(27), 3)

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

#### Raíz de una división (o cociente)
$$ 
\sqrt[n]{\frac{a}{b}} = \frac{\sqrt[n]{a}} {\sqrt[n]{b}}
$$


```python
expr_raw = r" \sqrt{\frac{4}{9}} "
result = sp.sqrt(sp.Integer(4) / sp.Integer(9))

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

```python
expr_raw = r" \sqrt[3]{\frac{1}{8}} "
result = sp.root(sp.Integer(1) / sp.Integer(8), 3)

display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(result)}"))
```

#### Raíz de números negativos
$$ 
\sqrt{-9} 
$$


No es un número real ya que:

$$3 \cdot 3 = 9$$

$$-3 \cdot -3 = 9$$

En general si a es un número negativo

$$ 
\sqrt[n]{a} 
$$

No es real si n es par

$$ 
\sqrt[n]{-a}  = -\sqrt[n]{a}
$$

Si  n es impar
