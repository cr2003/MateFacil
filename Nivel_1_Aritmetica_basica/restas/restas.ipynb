```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```
