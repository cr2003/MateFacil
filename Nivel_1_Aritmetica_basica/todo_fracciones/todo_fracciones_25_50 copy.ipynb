```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

```python
def frac_mixta_to_frac(entero, frac):
    """
    Recibe un entero y una fraccion y devuelve una fraccion impropia.
    frac.p es el numerador de la fraccion
    frac.q es el denominador de la fraccion
    """
    numerador = (frac.q * entero) + frac.p
    denominador = frac.q
    return sp.Rational(numerador, denominador)
```

```python
def frac_to_frac_mixta(frac):
    """
    Recibe una fraccion impropia y devuelve una cadena que representa una fraccion mixta
    """
    # Extraemos p y q
    p = frac.p
    q = frac.q

    # Calculamos la parte mixta
    entero = p // q
    residuo = p % q

    # Formateamos para mostrar
    if residuo == 0:
        resultado = rf"{entero}"
    else:
        resultado = rf"{entero} \frac{{{residuo}}}{{{q}}}"
    return resultado
```

###  Problema resuelto con multiplicación de fracciones

En un pueblo de 2,100 habitantes, la tercera parte es afectada por virus ¿Cuántos habitantes no tienen el virus?

$$\frac{1}{3} \times 2,100$$



```python
expr_raw = r"\frac{1}{3} \times 2,100"
frac = sp.Rational(1, 3) * 2_100
print("Habitantes afectados:")

display(Math(rf"{expr_raw} {spacing_equal} {fmt(sp.latex(frac))}"))
print("Habitantes sin virus:")
display(Math(rf"{fmt(2_100)} - {frac}{spacing_equal} {fmt(2_100 - frac)}"))
```
