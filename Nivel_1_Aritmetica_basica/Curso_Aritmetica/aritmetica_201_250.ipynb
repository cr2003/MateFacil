## Video 207

### Cómo calcular el IVA

CASO 1: Precio sin IVA

Una pantalla tiene un precio de $10 000 (sin incluir el IVA)

¿Cuánto se pagará de IVA?

$$\text{IVA = }16\% \text{  de  } 10\,000$$

```python
expr_raw = r"\text{IVA = }16\% \text{  de  } 10\,000"
precio = 10_000
porcentaje = 16
decimal = porcentaje / 100
IVA = int(decimal * precio)
total = precio + IVA
display(Math(rf"{expr_raw} {spacing} {sp.latex(IVA)}"))
display(
    Math(
        rf"\text{{total = Precio + IVA = }}{precio} + {IVA} {spacing} \${sp.latex(total)}"
    )
)
```

### Cómo calcular el IVA

CASO 2: Precio IVA incluido

Una pantalla tiene un precio de $11 600 (IVA incluido)

¿Cuánto se pagará de IVA?

$$\text{{Precio sin IVA = }} 11\,600 \div 1.16$$

```python
expr_raw = r"\text{{Precio sin IVA = }} 11\,600 \div 1.16"
total = 11_600
porcentaje = 16
decimal = 1 + (porcentaje / 100)
precio_neto = int(precio / decimal)

IVA = precio - precio_neto
display(Math(rf"{expr_raw} {spacing} {sp.latex(precio_neto)}"))
display(
    Math(
        rf"\text{{IVA = TOTAL - precio sin IVA}} {spacing} {sp.latex(total)}- {precio_neto} = \${IVA}"
    )
)
```

Cómo calcular el IVA

Juan compró una computadora de $15 000 (IVA incluido)

¿Cuánto pagó de IVA?

$$\text{{Precio sin IVA = }} 15\,000 \div 1.16$$

```python
expr_raw = r"\text{{Precio sin IVA = }} 15\,000 \div 1.16"
total = 15_000
porcentaje = 16
decimal = 1 + (porcentaje / 100)
precio_neto = float(round(total / decimal, 2))

IVA = float(round(total - precio_neto, 2))
display(Math(rf"{expr_raw} {spacing} {sp.latex(precio_neto)}"))
display(
    Math(
        rf"\text{{IVA = TOTAL - precio sin IVA}} {spacing} {sp.latex(total)}- {precio_neto} = \${IVA}"
    )
)
```

## Video 208

### Cómo calcular un descuento

Cecilia compró una televisión que tenía un costo de $18 600, de los cuales le hicieron un descuento del 18%.

¿Cuánto pagó por la televisión?

$$18\% \text{  de  } \$18\;600$$

```python
expr_raw = r"18\% \text{  de  } \$18\;600"
costo = 18_600
porcentaje = 18
decimal = porcentaje / 100
descuento = int(decimal * costo)
precio = costo - descuento

display(Math(rf"\text{{Descuento = }}{expr_raw} {spacing} {sp.latex(descuento)}"))
display(
    Math(
        rf"\text{{Precio neto = costo - descuento = }}{costo} - {descuento} {spacing} \${sp.latex(precio)}"
    )
)
```

* MCM: 

    Se listan todos los factores de los dos números y se elije el de máximo exponente

* MCD: 

    Se listan todos los factores comunes a los dos números y se elije el de menor exponente

#### Desomponer en factores primos

```python
import sympy as sp
from IPython.display import display, Math


def descomponer_en_factores_latex(numero):
    # 1. Obtenemos el diccionario de factores primos (ej: {2: 3, 3: 2, 5: 1, 7: 1})
    factores = sp.factorint(numero)

    partes_latex = []

    # 2. Recorremos los factores ordenados de menor a mayor
    for primo in sorted(factores.keys()):
        exponente = factores[primo]

        # Si el exponente es 1, solo mostramos el número primo
        if exponente == 1:
            partes_latex.append(str(primo))
        # Si es mayor que 1, lo elevamos usando la sintaxis de LaTeX
        else:
            partes_latex.append(f"{primo}^{{{exponente}}}")

    # 3. Unimos todas las partes usando el punto centrado '\cdot' como multiplicador
    expresion_factores = r" \cdot ".join(partes_latex)

    return expresion_factores
```

```python
num = 2520

res = descomponer_en_factores_latex(num)
display(Math(rf"{num} = {res}"))
```

```python
num = 8316

res = descomponer_en_factores_latex(num)
display(Math(rf"{num} = {res}"))
```

```python
a = 2520
b = 8316

# Calculamos el MCD y el MCM
# lcm (Lowest Common Multiple) y gcd (Greatest Common Divisor)
mcd_resultado = sp.gcd(a, b)
mcm_resultado = sp.lcm(a, b)

print(f"El MCD de {a} y {b} es: {mcd_resultado}")  # Resultado: 252
print(f"El MCM de {a} y {b} es: {mcm_resultado}")  # Resultado: 83160
```
