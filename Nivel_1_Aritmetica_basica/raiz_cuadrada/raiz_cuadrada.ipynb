```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

## Video 5

###  Cómo calcular una RAIZ CUADRADA AGREGANDO PUNTO DECIMAL

$$\sqrt{31}$$

```python
expr_raw = r"\sqrt{31}"
num = 31
resultado = sp.sqrt(num).evalf()

display(
    Math(rf"{expr_raw} {spacing_equal} {resultado} {spacing_approx} {fmt(resultado)}")
)
```
