### Cómo convertir una fracción a número decimal 

$$\frac{3}{5}$$

```python
# expr_raw = r"\frac{3}{5}"
frac = sp.Rational(3, 5)
resultado = frac.n(1)

# display(Math(rf"{expr_raw} {spacing} {sp.latex(frac)} {spacing} {resultado}"))
display(Math(rf"{sp.latex(frac)} {spacing} {resultado}"))
```
