## Video 101

Suma y resta de fracciones con enteros 

$$3 + \frac{2}{9}$$

```python
expr_raw = r"3 + \frac{2}{9}"
frac = 3 + sp.Rational(2, 9)
# Extraemos p y q
p = frac.p
q = frac.q

# Calculamos la parte mixta
entero = p // q
residuo = p % q
spacing = r"\,= \,"

# Formateamos para mostrar
if residuo == 0:
    resultado = rf"{entero}"
else:
    if entero:
        resultado = rf"{entero} \frac{{{residuo}}}{{{q}}}"
    else:
        resultado = rf"\frac{{{residuo}}}{{{q}}}"
display(Math(rf"{expr_raw} {spacing} {sp.latex(frac)} {spacing} {resultado}"))
```
