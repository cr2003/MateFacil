$$\sqrt{9} = 3 \implies 3 \times 3 = 9$$

```python
expr_raw = r"\sqrt{9}"
expr_sp = sp.sqrt(9)
# spacing = r"\quad = \qquad"  # Ajusta la cantidad de \qquad según necesites
spacing = r"\,= \,"
# expr_sp_format = f"{expr_sp:.3f}"
expr_sp_format = f"{expr_sp}"

display(Math(rf"{expr_raw} {spacing} {sp.latex(expr_sp_format)} "))
```

## Video 75

**1. Fracciones impropias**

Una fracción es **impropia** cuando el **numerador** es mayor o igual que el **denominador**. Representa una cantidad igual o mayor que un numero completo.

**2. Fracciones Mixtas**

Una fracción es **mixta** cuando está compuesta por un **numero entero** y una  **fracción propia** (una fracción donde el numerador es menor que el denominador).

$$\frac{5}{2} = 2 \frac{1}{2}$$

```python
expr_raw = r"\frac{5}{2}"
frac = sp.Rational(5, 2)

# Extraemos p y q
p = frac.p
q = frac.q

# Calculamos la parte mixta
entero = p // q
residuo = p % q

# Formateamos para mostrar
if residuo == 0:
    resultado = rf"{entero}"
else:
    resultado = rf"{entero} \frac{{{residuo}}}{{{q}}}"
display(Math(rf"{expr_raw} {spacing} {resultado}"))
```

## Video 77

Convertir fracciones mixtas a impropias

$$5 \frac{1}{2}$$

```python
expr_raw = r"5\frac{1}{2}"
entero = 5
numerador = 1
denominador = 2

p = (entero * denominador) + numerador
q = denominador
frac = sp.Rational(p, q)

display(Math(rf"{expr_raw} {spacing} {sp.latex(frac)} "))
```
