### Cómo sacar mitad, tercia, quinta y novena (Criterios de divisibilidad)

* **Mitad (Divisibilidad por 2)**

    Un número es divisible por 2 (o tiene mitad) si es un número par.

    O lo que es lo mismo, si termina en 0, 2, 4, 6 u 8

    #### Ejemplos:


```python
def divisible_2(num) -> bool:
    """Return True if num is divisible by 2, False otherwise"""
    return num % 2 == 0
```

```python
num = 278
print(f"¿{num} es divisible por 2?: {divisible_2(num)}")
```

* **Tercera (Divisibilidad por 3)**

    Un número es divisible por 3 (o tiene tercera) si es 3, 6 o 9.

    O sí la suma de sus dígitos da como resultado 3, 6 o 9

```python
def divisible_3(num) -> bool:
    """Return True if num is divisible by 3, False otherwise"""
    return num % 3 == 0
```

## Video 2 - Cómo sacar séptima

* **Séptima (Divisibilidad por 7)**

    Para saber sí un número es divisible por 7 (o tiene séptima), se siguen los siguientes pasos:

    1. Se quita el último dígito al número y se duplica dicho dígito.

    2. Al número que queda después de haberle quitado el último dígito se le resta el número obtenido el el paso 1

#### Ejemplos:

```python
def divisible_7(num) -> bool:
    """Return True if num is divisible by 7, False otherwise"""
    return num % 7 == 0
```

```python
num = 91
print(f"¿{num} es divisible por 7?: {divisible_7(num)}")
```

```python
num = 455
print(f"¿{num} es divisible por 7?: {divisible_7(num)}")
```

```python
num = 834
print(f"¿{num} es divisible por 7?: {divisible_7(num)}")
```

### Criterio de divisibilidad por 11

Para saber sí un número es divisible por 11, se siguen los siguientes pasos:

1. Se suman los dígitos que están en posición par

2. Se suman los dígitos que están en posición impar

3. Restar los resultados obtenidos en los pasos 1 y 2

Si el resultado es divisible por 11, entonces el número en cuestión también lo es.

#### Ejemplos:


```python
def divisible_11(num) -> bool:
    """Return True if num is divisible by 11, False otherwise"""
    return num % 11 == 0
```

```python
num = 143
print(f"¿{num} es divisible por 11?: {divisible_11(num)}")
```

```python
num = 2827
print(f"¿{num} es divisible por 11?: {divisible_11(num)}")
```
