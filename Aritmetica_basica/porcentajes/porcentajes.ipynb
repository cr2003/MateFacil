```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing = r"\,= \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

## Video 1

### Convertir fracción a porcentaje y viceversa

$$\frac{3}{4}$$

```python
expr_raw = r"\frac{3}{4}"
frac = sp.Rational(3, 4)
decimal = frac.evalf()
porcentaje = int(decimal * 100)

display(Math(rf"{expr_raw} {spacing} {sp.latex(decimal)} = {sp.latex(porcentaje)}\%"))
```

## Video 8

### Cómo calcular el IVA sobre un precio

Enrique compró una casa que tenía un costo de $250 000, pero le aumentaron el IVA de 16% ¿Cuánto pagó?

$$\text{IVA de }16\%  \text{ sobre }\; \$250\,000$$

```python
expr_raw = r"\text{IVA de }16\%  \text{ sobre }\; \$250\,000"
precio = 250_000
IVA_percent = 16
decimal = IVA_percent / 100
IVA = decimal * precio
IVA_fmt = fmt(IVA)

display(Math(rf"{expr_raw} {spacing} {sp.latex(IVA_fmt)}"))
display(
    Math(
        rf"Total {spacing} {fmt(amount)} + {sp.latex(IVA_fmt)} {spacing} \${fmt(amount + IVA)}"
    )
)
```

### ¿Cómo calcular el IVA? 

#### CASO 1: Precio sin IVA

Una pantalla tiene un precio de $10 000 (sin incluir el IVA) ¿Cuánto se pagará de IVA?

$$\text{IVA de }16\%  \text{ sobre }\; \$10\,000$$

```python
expr_raw = r"\text{IVA de }16\%  \text{ sobre }\; \$10\,000"
precio = 10_000
IVA_percent = 16
decimal = IVA_percent / 100
IVA = decimal * precio
IVA_fmt = fmt(IVA)

display(Math(rf"{expr_raw} {spacing} \${sp.latex(IVA_fmt)}"))
display(
    Math(
        rf"Total {spacing} {fmt(precio)} + {sp.latex(IVA_fmt)} {spacing} \${fmt(precio + IVA)}"
    )
)
```

#### CASO 2: Precio IVA incluido

Una pantalla tiene un precio de $11 600 (IVA incluido) ¿Cuánto se pagará de IVA?

$$\text{IVA de }16\%  \text{ incluido en total de}\; \$11\,600$$

```python
expr_raw = r"\text{IVA de }16\%  \text{ incluido en total de}\; \$11\,600"
total = 11_600
IVA_percent = 16
decimal = 1 + (IVA_percent / 100)
precio = total / decimal
IVA = total - precio
IVA_fmt = fmt(IVA)

display(Math(rf"{expr_raw} {spacing} \${sp.latex(IVA_fmt)}"))
# display(Math(rf"Total {spacing} {fmt(amount)} + {sp.latex(IVA_fmt)} {spacing} \${fmt(amount + IVA)}"))
```

## Video 10

### Cómo calcular un descuento

Cecilia compró una televisión que tenía un costo de $18 600, de los cuales le hicieron un descuento del 18%.

¿Cuánto pagó por la televisión?

$$18\% \text{  de  } \$18\;600$$

```python
expr_raw = r"18\% \text{  de  } \$18\;600"
costo = 18_600
porcentaje = 18
decimal = porcentaje / 100
descuento = decimal * costo
precio = costo - descuento

display(Math(rf"\text{{Descuento = }}{expr_raw} {spacing} {sp.latex(fmt(descuento))}"))
display(
    Math(
        rf"\text{{Precio neto = costo - descuento = }}{fmt(costo)} - {fmt(descuento)} {spacing} \${sp.latex(fmt(precio))}"
    )
)
```

## Video 11

### Problemas de Porcentajes y descuentos

#### a) 

Adrián compró un automóvil en $320 000, el precio incluía entre seguro, impuestos y accesorios 25% mas, ¿Cuál era el precio del automóvil sin contar seguro, impuestos y accesorios?

$$\text25\%  \text{ mas sobre }\; \$320\,000$$

```python
expr_raw = r"\text25\%  \text{ mas sobre }\; \$320\,000"
total = 320_000
additional_percent = 25
decimal = 1 + (additional_percent / 100)
precio = total / decimal
precio_fmt = fmt(precio)


display(
    Math(rf"{expr_raw}  \implies \text{{ precio neto: }} \$ {sp.latex(precio_fmt)}")
)
```

## Video 12

### Problemas de Porcentajes y descuentos

#### b) 

Paola compró una bicicleta de montaña en $1 200, si el precio incluia una rebaja de 20%, ¿cuál era el precio normal de la bicicleta?

$$\text20\%  \text{ descuento incluido en }\; \$1\,200$$

```python
expr_raw = r"\text20\%  \text{ descuento incluido en }\; \$1\,200"
total = 1_200
porcentaje = 20
decimal = 1 - (porcentaje / 100)
precio = total / decimal
precio_fmt = fmt(precio)


display(
    Math(rf"{expr_raw}  \implies \text{{ precio normal: }} \$ {sp.latex(precio_fmt)}")
)
```
