```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

```python
def frac_mixta_to_frac(entero, frac):
    """
    Recibe un entero y una fraccion y devuelve una fraccion impropia.
    frac.p es el numerador de la fraccion
    frac.q es el denominador de la fraccion
    """
    numerador = (frac.q * entero) + frac.p
    denominador = frac.q
    return sp.Rational(numerador, denominador)
```

```python
def frac_to_frac_mixta(frac):
    """
    Recibe una fraccion impropia y devuelve una cadena que representa una fraccion mixta
    """
    # Extraemos p y q
    p = frac.p
    q = frac.q

    # Calculamos la parte mixta
    entero = p // q
    residuo = p % q

    # Formateamos para mostrar
    if residuo == 0:
        resultado = rf"{entero}"
    else:
        if entero == 0:
            resultado = rf"\frac{{{residuo}}}{{{q}}}"
        else:
            resultado = rf"{entero} \frac{{{residuo}}}{{{q}}}"
    return resultado
```

```python
def decimal_a_fraccion(decimal_float):
    # 1. Convertimos el float a string para contar cuántos decimales tiene
    decimal_str = str(decimal_float).split(".")[-1]
    num_decimales = len(decimal_str)

    # 2. Calculamos el numerador y denominador de la fracción no simplificada
    numerador_original = int(round(decimal_float * (10**num_decimales)))
    denominador_original = 10**num_decimales

    # 3. Creamos las dos fracciones en SymPy
    # Usamos sp.sympify con la cadena exacta para evitar que se autoevalúe/simplifique
    frac_original = sp.sympify(
        f"{numerador_original}/{denominador_original}", evaluate=False
    )
    frac_simplificada = sp.Rational(numerador_original, denominador_original)

    # 4. Construimos la cadena en formato LaTeX
    latex_expr = (
        rf"{decimal_float} = "
        rf"{sp.latex(frac_original)} = "
        rf"{sp.latex(frac_simplificada)}"
    )

    # 5. Mostramos el resultado estético en el Notebook
    display(Math(latex_expr))
```

###  Fracciones con paréntesis: suma, resta, multiplicación y división

$$ \left( \frac{3}{2} - \frac{1}{4} \right) + \left( \frac{2}{5} - \frac{1}{2} \right) $$

```python
expr_raw = r" \left( \frac{3}{2} - \frac{1}{4} \right) + \left( \frac{2}{5} - \frac{1}{2} \right)"
frac = (sp.Rational(3, 2) - sp.Rational(1, 4)) + (sp.Rational(2, 5) - sp.Rational(1, 2))
display(Math(rf"{expr_raw} {spacing_equal} {sp.latex(frac)}"))
```
