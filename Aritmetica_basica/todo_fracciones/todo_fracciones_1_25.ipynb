```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing_equal = r"\,= \,"
spacing_approx = r"\, \approx \,"
```

#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

```python
def frac_mixta_to_frac(entero, frac):
    """
    Recibe un entero y una fraccion y devuelve una fraccion impropia.
    frac.p es el numerador de la fraccion
    frac.q es el denominador de la fraccion
    """
    numerador = (frac.q * entero) + frac.p
    denominador = frac.q
    return sp.Rational(numerador, denominador)
```

```python
def frac_to_frac_mixta(frac):
    """
    Recibe una fraccion impropia y devuelve una cadena que representa una fraccion mixta
    """
    # Extraemos p y q
    p = frac.p
    q = frac.q

    # Calculamos la parte mixta
    entero = p // q
    residuo = p % q

    # Formateamos para mostrar
    if residuo == 0:
        resultado = rf"{entero}"
    else:
        resultado = rf"{entero} \frac{{{residuo}}}{{{q}}}"
    return resultado
```

## Video 3

### Fracciones mixtas y fracciones impropias

* **Fracción impropia**

    El numerador es mayor que el denominador, ejemplo $\frac{5}{2}$
    
* **Fracción mixta**

    Las fracciones impropias se pueden representar como fracciones mixtas (parte entera y parte fraccionaria), ejemplo $2\frac{1}{2}$

## Video 5

### Cómo convertir fracciones mixtas en fracciones impropias

$$5\frac{1}{2}$$

```python
expr_raw = r"5\frac{1}{2}"
entero = 5
frac = sp.Rational(1, 2)
display(
    Math(rf"{expr_raw} {spacing_equal} {sp.latex(frac_mixta_to_frac(entero, frac))}")
)
```
