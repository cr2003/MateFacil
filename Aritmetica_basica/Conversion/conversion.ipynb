## Video 31
Convertir $150^\circ$ radianes

```python
angle_deg = 150 * u.deg
angle_rad = angle_deg.to(u.rad)

print(f"{angle_deg} = {angle_rad:.3f}")
```

```python
from sympy import pi, rad

# 1. Define the angle symbolically
angle_deg = 150

# 2. Convert to radians exactly
# Using rad() or multiplying by pi/180
angle_rad = rad(angle_deg)

# spacing = r"\quad = \quad"  # Ajusta la cantidad de \qquad según necesites
spacing = r"\, = \,"  # Ajusta la cantidad de \qquad según necesites


display(
    Math(
        rf" {sp.latex(angle_deg)} ^\circ {spacing} {sp.latex(angle_rad)} \text{{ rad}}"
    )
)
```

## Video 32
Convertir 11.3 $g/cm^3$ a $kg/m^3$

```python
densidad_gcm3 = 11.3 * (u.g / u.cm**3)
densidad_kgm3 = densidad_gcm3.to(u.kg / u.m**3)

print(f"{densidad_gcm3} = {densidad_kgm3:.0f}")
```

## Video 33
