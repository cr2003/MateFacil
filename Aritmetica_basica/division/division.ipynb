## Video 1

### Cómo dividir 

$$3396 \div 4 = 3396 / 4 = \frac{3396}{4} $$

```python
from astropy import units as u
import sympy as sp
from IPython.display import display, Math

a, m, x, y, z = sp.symbols("a, m, x, y, z", real=True)

spacing = r"\,= \,"
```

```python
expr_raw = r"\frac{3396}{4}"
frac = sp.Rational(3396, 4)
resultado = f"{frac:.2f}"


display(Math(rf"{expr_raw} {spacing} {sp.latex(frac)} {spacing} {sp.latex(resultado)}"))
```
